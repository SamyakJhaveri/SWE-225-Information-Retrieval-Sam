Read an HTML from a website
Trying to Parse price of an Iron Man Funko from https://www.funko.com/shop/details/marvel-avengers-endgame-18-inch-iron-man-pop-mega

```python
import requests
url = "https://www.newegg.com/p/037-00CJ-00130?Description=bikini&cm_re=bikini-_-037-00CJ-00130-_-Product"
result = requests.get(url) 
doc = BeautifulSoup(result.text, "html.parser")
print(result.text)
```

```python
doc = BeautifulSoup(result.text, "html.parser")
print(doc.prettify())
```

```python
prices = doc.find_all(text = "$")
parent = prices[0].parent
strong = parent.find("strong")
print(strong.string)
```

```python
tags = doc.find_all(["p", "div", "li"])
print(tags)
```

```python
tag = doc.find_all(["option"], text = "Undergraduate", value = "undergraduate")
print(tag)
```

Find Class

```python
tags = doc.find_all(class_ = "btn-item")
print(tags)
```

```python
# Regular expression 
import re
tags = doc.find_all(text = re.compile("\$.*."))
print(tags)
```
